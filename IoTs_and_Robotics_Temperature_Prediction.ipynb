```python
import pandas as pd
```

```python
train = pd.read_csv('/content/IOT_Train.csv')
train
```

```python
test = pd.read_csv('/content/IOT_Test.csv')
test = test.drop('id',axis=1)
```

```python
train=train.drop(['tambon_code','tambon_namt','amphur_code','amphur_namt','province_code','province_namt','latitude','longitude'],axis=1)
```

```python
test = test.drop(['tambon_code','tambon_namt','amphur_code','amphur_namt','province_code','province_namt','latitude','longitude'],axis=1)
```

```python
train['pm10']=train['pm10'].fillna(0)
train['pm2.5']=train['pm2.5'].fillna(0)
```

```python
test['pm10']=test['pm10'].fillna(0)
test['pm2.5']=test['pm2.5'].fillna(0)
```

```python
train[['date', 'time']] = train['time'].str.split(' ', expand=True)
```

```python
test[['date', 'time']] = test['time'].str.split(' ', expand=True)
```

```python
train[['time', 'timezone']] = train['time'].str.split('+', expand=True)
```

```python
train=train.drop(['timezone'],axis=1)
```

```python
test[['time', 'timezone']] = test['time'].str.split('+', expand=True)
test=test.drop(['timezone'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train.drop('temp',axis=1), train['temp'], test_size=0.2, random_state=42)
```

```python
cat_features = X_train.select_dtypes(include=['object']).columns.tolist()
model = CatBoostRegressor(cat_features=cat_features)
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_val, model.predict(X_val))
```
